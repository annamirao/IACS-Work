```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d
```

```python
mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')

mean = [10, 10]
cov = [[100, 50], [50, 100]]  # diagonal covariance
x, y = np.random.multivariate_normal(mean, cov, 10).T

n = 10
z = np.random.normal(loc=0.0, scale=1, size=n)

ax.scatter(x, y, z, label='test')

x_eval = np.linspace(-200,200,1000)
y_eval = np.linspace(-200,200,1000)

f = interp2d(x, y, z, kind='cubic')
z_new = f(x_eval, y_eval)
ax.plot_surface(x_eval, y_eval, f(x_eval, y_eval), label='hello')

#ax.legend()

plt.show()
```

```python
mean = [10, 10]
cov = [[100, 50], [50, 100]]  # diagonal covariance
x, y = np.random.multivariate_normal(mean, cov, 1000).T

heatmap, xedges, yedges = np.histogram2d(x, y, bins=50)
extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

plt.clf()
plt.imshow(heatmap.T, extent=extent, origin='lower')
```
