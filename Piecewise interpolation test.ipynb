```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import splrep, splev
%matplotlib inline
```

```python
plt.figure(figsize=(16,8))
# pick random points
n = 15
xdata = np.linspace(-200,200,n)
ydata = np.random.normal(loc=0.0, scale=1, size=n)

plt.plot(xdata, ydata, 'ro', label='Data')

eval_points = np.linspace(-200,200,1000)

# x[i] <= xp   < x[i+1]
# S = sum(c[m, i] * (xp - x[i])**(k-m) for m in range(k+1))

arr = np.array([-75, 0, 75])

options = (1, 2, 3, 4, 5)

# b = (-100.0)
# e = (100.0)

for o in options:
    tck = splrep(xdata, ydata, xb=None, xe=None, k=o, t=arr)
    plt.plot(eval_points, splev(eval_points, tck, der=0), label='degree ' + str(o) + ' piecewise')

plt.legend()
```
